```python

from SDMXtoV4.SDMXtoV4 import SDMXtoV4, SDMXsummary

infile = "NATP.ESA10.0302_SDMX Output_BlueBook_29_Sep_2017.xml"

# not taking reference year price as its metadata
# not taking stock as later changes reduce it to a one item dimension (i.e metadata)
df = SDMXtoV4(infile, ["activity","instr_asset","prices"], returnFrame=True)

SDMXsummary(infile)

"""
Droppping the following

{'activity': ['Total - All activities'], 
'instr_asset': ['Cultivated biological resources (gross)', 'Inventories by type of inventory (gross)']}
"""

df = df[df["activity"] != "Total - All activities"]
df = df[df["instr_asset"] != "Cultivated biological resources (gross)"]
df = df[df["instr_asset"] != "Inventories by type of inventory (gross)"]

# tidy names up
colMap = { 
    "Time_codelist":"yyyy", 
    "Time":"time", 
    "Geography_codelist":"geography_codelist", 
    "Geography":"geography",
    "instr_asset":"instrument-asset",
    "ESTAT.CL_ACTIVITY.1_1":"business-investment-activity",
    "ESTAT.CL_INSTR_ASSET.1_1":"business-investment-instrument-asset",
    "ESTAT.CL_NA_PRICES.1_1":"business-investment-prices",
    "ESTAT.CL_NA_STO.1_1":"stocks",
    "sto":"stocks"
}

newCols = []
for col in df.columns.values:
    
    if col in colMap.keys():
        col = colMap[col]
    newCols.append(col)
        
df.columns = newCols

df["yyyy"] = df["time"]

df.to_csv("v4_buisInvest_0302.csv", index=False)
```
